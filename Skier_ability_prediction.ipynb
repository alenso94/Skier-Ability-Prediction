Metrics which do not seem to have normal distributions by visualizing the distributions are given below: m1, m2, m4, m5, m6, m8, m9, m10, m14.

Let's perform Shapiro-Wilk test evaluate the same. 

The Shapiro-Wilk test evaluates a data sample and quantifies how likely it is that the data was drawn from a normal distribution. 

The test assume that that the sample was drawn from a normal distribution. This is called the null hypothesis, or H0. A threshold level is chosen called alpha, typically 5% (or 0.05), that is used to interpret the p-value. Here I have chosen alpha value as 4%(0.04).

p <= alpha: reject H0, not normal.
p > alpha: fail to reject H0, normal.

I am seeking results with a larger p-value to confirm that the sample was likely drawn from a normal distribution.

A result above 4% does not mean that the null hypothesis is true. It means that it is very likely true given available evidence.    

```python
#import libraries to perform Shapiro-Wilk test
from scipy.stats import shapiro
metrics = ['m1','m2','m3','m4','m5','m6','m7','m8','m9','m10','m11','m12','m13','m14']
#metrics with non normal distribution
non_norm=[]
#check if each metric has normal distribution
for m in metrics:
    print (m+":")
    #shapiro-wilk test
    stat,p = shapiro(df[m])
    #p-value
    print(p)
    #alpha is set as 0.04
    alpha = 0.04
    if p > alpha:
        print(m+' looks Normal (fail to reject H0)')
    else:
        print(m+' does not look Normal (reject H0)')
        non_norm.append(m)
print(non_norm)
```

After performing Shakiro-Wilk test, metrics such as m1, m2, m3, m4, m5, m6, m8, m9, m10, m11, m12, m14 do not have normal distribution. 

```python
#preparing the features and the label
#training features
x = df.drop('ability', axis=1)
#training label
y = df.ability
```

```python
# Scaling all the features using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = pd.DataFrame(scaler.fit_transform(x), columns=list(x.columns))
```

```python
#splitting data into train and test set in the ratio 8:2
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
```

```python
# ANOVA feature selection
from sklearn.feature_selection import SelectKBest, f_classif
# configure to select all features
fs = SelectKBest(score_func=f_classif, k='all')
# learn relationship from training data
fs.fit(x_train, y_train)
```

```python
from matplotlib import pyplot
# scores for the features
for i in range(len(fs.scores_)):
    print('Feature %d: %f' % (i, fs.scores_[i]))
# plot the scores
pyplot.bar([i for i in range(len(fs.scores_))], fs.scores_)
pyplot.show()
```

This shows that there are 5 features with comparitively less scores(<50). So, we can set k as 9. 

```python
# define feature selection(9 best features)
fs = SelectKBest(score_func=f_classif, k=9)
# learn relationship from training data
fs.fit(x_train, y_train)
# apply feature selection to training and test data
x_train = fs.transform(x_train)
x_test = fs.transform(x_test)
```

As obtained from the scores, the features selected are m2, m3, m4, m5, m8, m9, m10, m11, m13 and the features removed are m1,m6,m7,m12,m14

Decision tree can be used as an interpretable model

```python
#fit decision tree model
from sklearn import tree
dtree = tree.DecisionTreeClassifier()
dtree.fit(x_train,y_train)
```

```python
#import libraries for metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
#make predictions using the train set
y_pred_train=dtree.predict(x_train)
#training accuracy
accuracy_train= accuracy_score(y_train, y_pred_train)
print("Training Accuracy ",accuracy_train*100)
#make predictions using the test set
y_pred_test =dtree.predict(x_test)
#print the classification report
print(classification_report(y_test, y_pred_test))
#testing accuracy
accuracy_test= accuracy_score(y_test, y_pred_test)
print("Testing Accuracy ",accuracy_test*100 )
print("Confusion matrix")
#confusion matrix
cfm = confusion_matrix(y_test, y_pred_test)
print(cfm)
```

```python
#check the value counts to identify the label
y_test.value_counts()
```

```python
#function to create confusion matrix
# Normalise
cmn = cfm.astype('float') / cfm.sum(axis=1)[:, np.newaxis]
fig, ax = plt.subplots(figsize=(7,7))
sns.heatmap(cmn, annot=True, cmap="YlGnBu", fmt='.2f', xticklabels=['advanced', 'beginner', 'expert', 'intermediate'], yticklabels=['advanced', 'beginner', 'expert', 'intermediate'])
plt.ylabel('Actual')
plt.xlabel('Predicted')
```

```python
# save the model to disk
import pickle
filename = 'finalized_model.sav'
pickle.dump(dtree, open(filename, 'wb'))
```
